```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/IRIS.csv')
df.head()
```

```python
#Boxplot
sns.boxplot(data=df)
plt.show()
```

```python
#to generate heatmap
#generate a heatmap using the above data
import matplotlib.pyplot as plt
import seaborn as sns

# Select only the numerical features for correlation calculation
numerical_features = df.select_dtypes(include=['number'])

# Generate the heatmap with the 'RdBu' colormap (red to blue)
sns.heatmap(numerical_features.corr(), annot=True, cmap ='coolwarm', linewidths=0.5)
plt.show() # Add this line to display the heatmap
```

```python
#violin plot
sns.violinplot(data=df)
plt.show()
```

```python
#pairplot
sns.pairplot(df)
plt.show()
```

```python
#scatter plot
sns.scatterplot(data=df)
plt.show()
```

```python
#do level encoding
label_encoder = LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])
df.head()
print(df)
```

```python
#now create heatmap again using encoded level
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame with the level-encoded 'species' column
# Select all numerical features, including the encoded 'species' column
numerical_features = df.select_dtypes(include=['number'])

# Generate the heatmap
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
